```python
import requests
from bs4 import BeautifulSoup
import pymysql
import time
```

```python
job_title = '%E8%BB%9F%E9%AB%94%E8%A8%AD%E8%A8%88%E5%B7%A5%E7%A8%8B%E5%B8%AB' # 軟體設計工程師 (使用關鍵字查詢)
resp = requests.get("http://www.104.com.tw/i/apis/jobsearch.cfm?kws=" + job_title + "&fmt=8&cols=JOB")
url_template = "http://www.104.com.tw/i/apis/jobsearch.cfm?kws=" + job_title + "&fmt=8&cols=JOB,NAME,PCSKILL_ALL_DESC,ADDR_NO_DESCRIPT,ADDRESS&PAGE={}"
total_pages = resp.json()["TOTALPAGE"]
pages = int(total_pages) + 1
print('Total pages: ' + total_pages)

conn = pymysql.connect(host='localhost', port=3306, user='root', password='1234', db='jobs', use_unicode=True, charset="utf8mb4", autocommit=True)
cur = conn.cursor()
insert_104_template = "insert into job1(site, company, title, location, skills) values(104, %s, %s, %s, %s)"

for page in range(1, pages):
    time.sleep(1) # 每秒爬一次避免還未resp回來程式就往下跑
    print('page: ' + str(page))
    url = url_template.format(page)
    
    try:
        resp = requests.get(url, timeout=5)
        if resp is not None and 'data' in resp.json():
            datas = resp.json()['data']
        else:
            continue
    except:
        continue

    # 若沒該欄位則給空值
    for data in datas:
        if 'PCSKILL_ALL_DESC' not in data:
            data['PCSKILL_ALL_DESC'] = ''
        if 'ADDR_NO_DESCRIPT' not in data:
            data['ADDR_NO_DESCRIPT'] = ''
        if 'ADDRESS' not in data:
            data['ADDRESS'] = ''

    jobs = [(data['NAME'], data['JOB'], data['ADDR_NO_DESCRIPT']+data['ADDRESS'], data['PCSKILL_ALL_DESC']) for data in datas]
    cur.executemany(insert_104_template, jobs)
    
cur.close
conn.close
```
